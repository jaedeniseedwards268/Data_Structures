The coeffients are a, b and c.
The variables a = 1, b=1, c=-6






```python
import matplotlib.pyplot as plt

a = 1
b = 1
c = -6

x_values = [-2, 0, 2.1]
for x in x_values:
    y = a * x**2 + b * x + c
    print(f'y evaluated at x = {x} is {y}')

```

The floating point number displayed for the evaluation of y has a lot of places after the decimal point.

The following code will limit the places to 3 making the output look more comprehensible and save memory space.

print(f'{value:.3f}')

Plotting and visualization of data is important to give a visual representation of your code.

```python
import numpy as np
x = np.linspace(-4, 4, 100)
y = a * x**2 + b * x + c
plt.plot(x, y)
plt.title('Polynomial Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()

```

Exercise 3: Adding descriptive labels, tick marks on the x and y axis and grid lines will make a better visual reference and better visibility.

1. Add a legend and tick marks with the following code.  
2. Make sure you replace the necessary code for your chart that you created in exercise 2.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-3, 3, 100)
# Calculate y values based on the polynomial
y = (x + 2) * (x - 1) * (x - 2)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the polynomial
plt.plot(x, y, 'r--', label='y = (x + 2)(x - 1)(x - 2)')  # Line plot with legend

# Plot points where y equals zero
plt.plot([-2, 1, 2], [0, 0, 0], 'bo', markersize=10, label='Roots (y=0)')  # Points on the x-axis

# Set labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Second Nice Python Figure of Your Name')

# Customize tick marks
plt.xticks(ticks=np.arange(-3, 4, 1))  # X-axis ticks from -3 to 3
plt.yticks(ticks=np.arange(-10, 11, 2))  # Y-axis ticks from -10 to 10

# Add a legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
