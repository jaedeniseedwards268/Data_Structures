```python
import matplotlib.pyplot as plt
import seaborn as sns
import random

#Generate random data
data = [random.randint(1,100) for __ in range(100)]

#Create a histogram
sns.histplot(data, bins=20)
plt.title("Distribution of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Create a program that fetches historical stock price data from a financial API and visualizes it using a line chart.

Import necessary libraries:
1. Import the `pandas`, `matplotlib.pyplot`, and `yfinance` libraries.

2. Fetch stock price data:
   Use the `yfinance` library to fetch historical stock price data for a specific ticker symbol (e.g., 'AAPL' for Apple).
  Specify the start and end dates for the data.

3.Prepare the data:
   Extract the relevant columns (e.g., 'Date', 'Open', 'High', 'Low', 'Close') from the fetched data convert the 'Date' column to a datetime format.
4. Create a line chart:**
   Use the `plot()` function from `matplotlib.pyplot` to create a line chart.
   Customize the chart with labels, titles, and formatting options.

5. Display the chart:
  Use the `plt.show()` function to display the chart.



```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch stock price data
ticker = 'AAPL'
start_date = '2023-01-01'
end_date = '2024-10-18'

data = yf.download(ticker, start=start_date, end=end_date)

# Prepare the data
data.reset_index(inplace=True)
data['Date'] = pd.to_datetime(data['Date'])

# Create a line chart
plt.plot(data['Date'], data['Close'])
plt.title(f"Closing Price of {ticker}")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
